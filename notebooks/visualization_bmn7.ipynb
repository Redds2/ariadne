```python
def print_constraints_station(df, station_id):
    station = df[df.station == station_id]
    min_x, max_x = station.x.min(), station.x.max() 
    min_y, max_y = station.y.min(), station.y.max() 
    min_z, max_z = station.z.min(), station.z.max()
    print("Station %d:" % station_id)
    print("min_x, max_x: [%f, %f]" % (min_x, max_x))
    print("min_y, max_y: [%f, %f]" % (min_y, max_y))
    print("min_z, max_z: [%f, %f]" % (min_z, max_z))

for stat in data.station.unique():
    print_constraints_station(data, stat)

print("Global:")
print("min_x, max_x: [%f, %f]" % ( data.x.min(), data.x.max() ))
print("min_y, max_y: [%f, %f]" % (data.y.min(), data.y.max()))
print("min_z, max_z: [%f, %f]" % (data.z.min(), data.z.max()))
```

```python
event = data[data.event == 0].copy()
event.loc[event.det == 1, 'station'] = event.loc[event.det == 1, 'station'].values + 3
event
```

```python
from matplotlib import collections  as mc



def visualize_2d_points_from_out( x, y, col, show=False, ax_in=None, scatter_kwargs=None):    
    if ax_in is None:
        fig, ax = plt.subplots(figsize=(8, 6))
    else:
        ax = ax_in
    scatter_kwargs = {} if scatter_kwargs is None else scatter_kwargs
    ax.scatter(x, y, c=col, **scatter_kwargs)
    if show:
        plt.tight_layout()
        plt.show()
        return
    return ax
```

```python

true_points = event[event.track != -1]
false_points = event[event.track == -1]

ax = visualize_2d_points_from_out(true_points.x,true_points.z, 
                                  'tab:green', 
                                  scatter_kwargs={'zorder':1, 's':15})
visualize_2d_points_from_out(false_points.x,false_points.z, [0.1, 0.1, 0.1, 1], 
                             show=True, ax_in=ax, 
                             scatter_kwargs={'zorder':2, 's':4})
```
